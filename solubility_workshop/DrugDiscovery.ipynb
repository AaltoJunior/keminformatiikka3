# Keminformatiikan historia

1960-luvulla laskennalliset menetelmät luonnontieteissä, kuten biologiassa alkoivat lisääntymään, ja erityisesti Corwin Hanschia ja hänen kollegoitaan voidaan kiittää ensimmäisistä tutkimuksista, joissa huomattiin yhdisteiden ja biologisten vaikutusten perustuvan yhdisteiden rakenteeseen.

Hansch ja muut huomasivat, että tutkittavan yhdisteen pienetkin rakenteelliset muutokset aiheuttivat kasvin auksiinin (kasvuhormoni) toiminnan ja vaikutuksen korostumista. He siis löysivät yhdisteitä, jotka olivat **biologisesti aktiivisia** eli aiheuttivat biologisen vasteen. 

Tämä aineen rakenteen suhde biologiaan oli matemaattisesti todettavissa korrelaationa, eli muuttujien välisenä riippuvuutena. Lisäksi aineen pitoisuus vaikutti myös biologiseen vasteeseen usein suoraan verrannollisesti, eli positiivisesti korreloiden. Näistä tuloksista **syntyi teoria rakenteen ja aktiivisuuden yhteydestä**, joka voitiin ilmaista matemaattisena kaavana. Tuloksista syntyi siis malli, jolla voidaan ennustaa aineen ominaisuuksia ennen kokeellisia varmentavia tutkimuksia. 

Hansch nimesi tämän mallin kvantitatiiviseksi rakenne-aktiivisuussuhteeksi (engl. _Quantitative Structure-activity-relationship_, QSAR). 



**Kuva 1.** QSAR-malleja käytetään ennustamaan rakenteen perusteella ominaisuuksia esimerkiksi lääketutkimuksia varten. 

Tutkijat innostuivat Hanschin tuloksista tutkimaan myös muita tapauksia, kuin auksiinin aktiivisuuteen vaikuttavia yhdisteitä. Kävi ilmi, että laaja kirjo lääkkeitä ja luonnonyhdisteitä vaikutti rakenteellisesti johdetuilla laskennallisilla menetelmillä biologiseen aktiivisuuteen.

Nämä mallit johdettiin datasta, mutta sitä syntyi kuitenkin niin paljon, että vain tietokoneilla aineistoja voitiin käsitellä dataa kustannustehokkaasti. Täten 1960–1970-luvulla kemistit ja tietojenkäsittelytieteilijät alkoivat  tutkimaan miten kemiallista dataa voidaan parhaiten esittää ja varastoida tietoteknisesti.

## Yhdisteiden esittäminen tietokoneille: SMILES-merkkijono

SMILES (Simplified Molecular Input Line Entry System) on osoittautunut luotettavan yksinkertaiseksi tavaksi viestiä **molekyylirakenteesta** tietokoneelle. Lisäksi sen yksinkertainen tiedostotyyppi, joka koostuu kirjainten ja numeroiden jonosta tekee siitä niin ihmiselle kuin tietokoneelle helpon tavan kirjoittaa, lukea ja tallentaa rakennekaavojen kuvauksia. Tämä oli myös merkittävä etu aikana, jolloin tietokoneiden muistikapasiteetti oli hyvin rajallinen.

**SMILES koostuu kirjainten ja numeroiden sarjasta**, jota kutsutaan *merkkijonoksi*. Se luetaan tietyllä säännöillä ja algoritmeilla, jonka tarkoitus on muodostaa graafi, eli pisteiden ja janojen verkko. Verkko kuvaa atomeja niin kutsuttuina solmuina ja niiden välisiä sidoksia viivoina. Yhdessä nämä muodostavat verkon, jota tietokone lukee matemaattisesti verkkoteorian säännöillä. Tässä työpajassa emme syvenny näihin algoritmeihin ajan puutteen vuoksi.

Piirto-ohjelmistoilla piirtämäsi molekyylit voidaan esittää SMILES-merkkijonona, ja esimerkiksi **MarvinSketch voi tallentaa ja tuoda tässä formaatissa molekyylejä.**


### Säännöt SMILES-merkkijonolle

Alla sijaitsevassa kuvassa on esimerkki lääkeyhdisteestä, jonka rakenne on ilmaistu SMILES-merkkijonona. Väritetty osat, tai fragmentit, on kuvattu tietyllä tavalla hyödyntäen vakiintuneita sääntöjä, mitä algoritmisesti tietokone tulkitsee aina tietyllä tavalla riippuen merkkien järjestyksestä. Kyseinen yhdiste tunnetaan nimellä: vaniliini tai 4-hydroksi-3-metoksibentsaldehydi.



**Kuva 3.** SMILES-merkkijonossa rakenteita voidaan eritellä vastaamaan tiettyjä atomeja ja niiden välisiä sidoksia tietyillä säännöillä.

Säännöt SMILES-merkkijonojen lukemiseen ei valitettavasti tavanomainen molekyylikaavojen lukemisen tulkinta riitä. Alla oleva taulukko kuvaa, miten SMILES-merkkijonoa tietokone on tulkinnut.

**Taulukko 1.** SMILES-säännöt erilaisilla esimerkeillä.
| Symboli | Esim. | Selitys |
| --- | --- | --- |
| Kemiallinen merkki | O | Kyseessä on molekyyli, vesi. Alkuaineet ilmaistaan kemiallisina merkkeinä, kuten jaksollisessa järjestelmässä. Vetyjä ei ilmoiteta vaan ne tulkitaan automaattisesti osaksi yhdistettä perustuen alkuaineen tyypilliseen valenssiin, eli sidosten lukumäärään. Katso koodista miten ne saadaan näkyviin |
| Kovalenttinen sidos | C=C | Yksinkertaisia sidoksia ei esitetä, mutta kaksois- ja kolmoissidokset esitetään = ja # symbolein |
| Vedyt | C | Merkitään vain kiraalisuuden määrittämisessä, C vastaa metaania eli CH4:sta |
| Haarat | CCCC(C)CC | Sulut () ilmoittavat haaran ja suluissa ilmoitetaan mikä haara on kyseessä  |
| Rengasrakenne | C1CCCCC1 tai C1CCC(C(C1)Cl)Cl | Numerot 1-9 ilmoittavat atomin sijainnin rengasrakenteessa. Vedyt korvanneet substituentit voidaan laittaa sulkuihin. HUOM. Cl on kloori-atomi|
| Aromaattisuus | c1ccccc1 | Rengasrakenteen kiraalisuus voidaan esittää vaihtamalla kirjaimet pieniksi |
| Kiraalisuus | C[C@@H](C(=O)O)N  | L-alaniini, jossa @@ hiilen jälkeen osoittaa sen kiraaliseksi |
| Isotoopit | [13C] | C13-isotoopeista koostuva etaanimolekyyli. Numerot ennen alkuainemerkkiä osoittavat isotoopin |
| Sähköinen varaus ja radikaalit | [NH4+].[Cl-] ; C[CH2].[CH3]  | Varaus + tai -, ja radikaalit usein luetaan erityisillä algoritmeilla ohjelmistoissa, jotka tulkitsevat SMILES-merkkijonoja. Niille ei ole vakiintunutta merkintää  |


```python
# Testaa taulukon SMILES-merkkijonoja tällä koodisolulla. Samalla voit tutustua siihen, mitä esitietoja koodi tarvitsee.

from rdkit import Chem
from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG

smiles = 'O'  # Voit siirtää haluamasi SMILES-merkkjonon '-merkkien väliin. Paina lopuksi CTRL + ENTER
m = Chem.MolFromSmiles(smiles)
# m = Chem.AddHs(m) # Voit tällä komennolla visualisoida kaikki vedyt, jos haluat. Poista ensimmäinen #-merkki aktivoidaksesi asetuksen
def moltosvg(mol, molSize=(350, 300), kekulize=True):
    mc = Chem.Mol(mol.ToBinary())
    if kekulize:
        try:
            Chem.Kekulize(mc)
        except:
            mc = Chem.Mol(mol.ToBinary())
    if not mc.GetNumConformers():
        rdDepictor.Compute2DCoords(mc)
    drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0], molSize[1])
    drawer.DrawMolecule(mc)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    return svg.replace('svg:', '')

SVG(moltosvg(m))
```

## SMILES:n formaatit

Taulukon 1. SMILES-kaavat voidaan jakaa vielä kahteen tyyppiin. Yleensä nämä tyypit erottaa siitä, sisältääkö SMILES '@' tai '@@' merkkejä. 

Näitten SMILES-formaattien erona on niiden eri tarkoitus informatiikassa, jossa isomeeriset SMILES-merkkijonot kuvaavat kiraalisia yhdisteitä ja kanoniset tallennetaan ainutlaatuisina merkkijonoina tietokantoihin.

### Kanoninen SMILES:

- Kanoninen SMILES on ainutlaatuinen merkkijono molekyylirakenteesta.
- Ainutlaatuisuus mahdollistaa nimikkeiden erottelun tietokannoista SMILES merkkijonoilla
- Hyödyllinen erityisesti kun etsitään tarkkaa rakennetta mistä tahansa datasetistä, sillä variantteja ei ole.
- Tätä kannattaa hyödyntää aina, kun haetaan yhdisteitä tietokannoista.

### Isomeerinen SMILES:

- Nimensä mukaisesti ottaa huomioon yhdisteessä isomerian.
- Jos yhdiste on enantiomeeri, vain isomeerinen SMILES voi erottaa yhdisteet toisistaan, vaikka niillä on sama molekyylikaava.
- Tämä on lääkekehityksessä erityisen tärkeää, koska kolmiuloitteinen rakenne on erittäin selektiivinen biologiselta aktiivisuudeltaan (Esim. Talidomidi tapaukset perustuvat saman lääkkeen enantiomeerien suunnattomiin toksisuuseroihin erityisesti raskaina olevien naisten sikiöiden terveyden suhteen).

```python
# SMILES-merkkijonoja voidaan myös numeroida, mutta ne eivät noudata samoja sääntöjä, kuin yhdisteiden nimeämisessä.

from rdkit import Chem
from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG

def moltosvg_with_indices(mol, molSize=(350, 300), kekulize=True):
    mc = Chem.Mol(mol.ToBinary())
    if kekulize:
        try:
            Chem.Kekulize(mc)
        except:
            mc = Chem.Mol(mol.ToBinary())
    if not mc.GetNumConformers():
        rdDepictor.Compute2DCoords(mc)
    drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0], molSize[1])
    drawer.drawOptions().addAtomIndices = True  # Numerointi alkaa tietokoneelle numerosta 0
    drawer.DrawMolecule(mc)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    return svg.replace('svg:', '')

smiles = 'CCC(CC)CC'  # Syötä haluamasi SMILES tähän.
m = Chem.MolFromSmiles(smiles)

SVG(moltosvg_with_indices(m))
```

## Kemian data ja tietokannat

Kemiallisia aineistoja tallennetaan tietokantoihin, joista  tutkijat voivat myöhemmin hakea ja viedä aineistoja jatkotutkimuksia varten tai vain tarkistaa yhdisteen ominaisuuksia ennen laboratorioon suuntaamista.

## PUG-REST menetelmä ja SMILES

Ohjelmistoympäristöön voi tuoda helposti yhdisteitä teettämällä kyselyitä PubChem-palvelimelta. Tämä on erityisen hyödyllistä, kun halutaan nopeasti tuoda tietoa lääkeyhdisteistä ilman hakukoneita luettelemalla haluttuja molekyylejä osaksi koodia. 

Tästä on erityistä hyötyä, kun haluat hakea SMILES-merkkijonoja tietokannasta.

```python
# Tee lista tuntemistasi lääkemolekyyleistä tähän englannin kielellä ja erota jokainen nimike pilkulla ja lainausmerkeillä seuraavanlaisesti.

# Esimerkiksi, kirjoita englannin kielellä.
compound_names = ["Aspirin","Vinorelbin", "Venlafaxin", "Thalidomide", "Benzamide", "Cytosine"] 
```

```python
import requests
 # Älä muokkaa tätä osuutta koodista. 
def retrieve_names(compound_names):
    pugrest = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
    pugoper = "property/CanonicalSMILES"
    pugout = "txt"
    results = {}
    for compound_name in compound_names:
        pugin = "compound/name/" + compound_name
        url = f"{pugrest}/{pugin}/{pugoper}/{pugout}"
        response = requests.get(url)
        if response.status_code == 200:
            results[compound_name] = response.text.strip()  # Strip to remove leading/trailing whitespaces
        else:
            results[compound_name] = f"Error: {response.status_code}"
    return results

# Muokkaa alla olevaa koodiriviä.
compound_names = ["", ""] 
# Muokkaa vain yllä olevaa koodiriviä.
canonical_smiles = retrieve_names(compound_names)
for compound_name, smiles in canonical_smiles.items():
    print(f"Canonical SMILES for {compound_name}: {smiles}")
```

### Lääkeaineiden toiminnan perusta

Lääkeaineet ovat moninainen joukko. Ne perustuvat kuitenkin biologiseen aktiivisuuteen joko lisäämällä tai ehkäisemällä jotakin biokemiallista prosessia tai prosesseja kehossasi. ### Lääkeyhdisteiden ominaisuudet kehossa

Lääkeyhdisteiden vaikutuksia ja vaikutusten perustaa on ollut kuitenkin hankalaa tieteellisesti tutkia. 1800-luvulla oli tavanomaista nojata hyvin kokeelliseen lääkeyhdisteiden etsimiseen vapaaehtoisilla sairaiilla potilailla. Lääkärien toiminta perustui pitkälle rohtojen ja yrttien sekä muitten hoitojen perimätietoon, omaperäisiin tulkintoihin ja kokemukseen sekä väliajoin heidän työtään ohjasi esimiehen, usein tutkivan ylilääkärin auktoriteett. Vasta 1800-luvun puolivälissä kemialliset erotus- ja analyysimenetelmät sekä käsite funktionaalisista ryhmistä eli rakenteiden vaikutuksesta kemiallisiin ominaisuuksiin edisti ymmärrystä tiettyjen yhdisteiden vaikutuksista.

#### ADME-ominaisuudet 

Kemialtaan lääkeaineiden biokemialliset vaikutukset jaetaan neljään ryhmään: imeytyminen, jakautuminen, metabolia ja erittyminen. Nämä muodostavat englannin kiellä lyhenteen ADME (_Adsorbtion, distribution, metabolism and elimination_) 

Yksi merkityksellisimmistä ominaisuuksista on lääkkeen **imeytyminen**, eli miten nautittu aine imeytyy verenkiertoon. Keskeisimpiä keinoja ovat:

1. Oraalinen: nautitaan suun kautta, mutta kulkeutuminen kehoon voi tapahtua läpäisemällä joko suolistoepiteelin.  Helppo ja tyypillisin tapa saada kehoon lääkeainetta. Lisäksi maksan ja suoliston kyky muokata metabolisesti entsyymeillään lääkeyhdistettä ennen isoon verenkiertoon imeytymistä on merkityksellistä esim. venlafaksiinille.
2. Posken limakalvon tai kielenalaisen kudosepiteelin läpäiseminen: kudosepiteeli on erittäin ohut ja täten tämä on nopea keino imeyttää lääkeainetta. Se voi toisaalta olla haitallinen, sillä lääkeaineen pitoisuus nousee nopeasti potilaan veressä.
3. Peräsuoli: lääke saadaan verenkiertoon verisuonitiheän sulkijalihaksen limakalvon läpi imeytettyä.

Voimme päätellä että erityisesti limakalvot ovat merkityksellisiä imeytymisen kannalta, mutta ohuen kudosesteen lisäksi lääkeaineella täytyy olla tietty ominaisuus: **hyvä vesiliukoisuus**, sillä veri ja muutenkin kehomme on enimmäkseen vettä. Vesi on siis keskeisin liuotin kehossamme.

Muita lääkeyhdisteiden ominaisuuksia on **jakautuminen** kehossa, joka on mitta lääkepitoisuuden jakautumisesta elimissä ja kudoksissa. Tähän vaikuttaa hyvin moni asia, mutta keskeisimpiä tekijöitä ovat tiettyjen biologisten esteiden kuten aivo-veriesteen tai proteiinien ja rasvojen pitoisuuksien vaikutus. 

**Metabolia** on merkittävä askel imeytymisen ja jakautumisen jälkeen, kun keho kohtelee lääkeainetta kuona-aineena. Se pyritään poistamaan joko ulosteen tai virtsan mukana, mutta saattaa myös muokata lääkeainetta vesiliukoisemmaksi. Pääosin maksa ja sen entsyymit vastaavat tästä prosessista. Lääkeaineen tarkoitus toisaalta perustuu juurikin metaboliaan, eli kehon metaboliaa tutkimalla voidaan myös löytää lääkeaineita, jotka joko edistävät tai estävät joitain metabolisia prosesseja joko suoraan tai epäsuorasti.

**Erittyminen** on prosessi, jossa paksusuoli ja munuaiset poistavat lääkeainetta verenkierrosta. Tämä voi olla erityisen vaikeaa rasvaliukoisten yhdisteiden tai raskasmetalleja sisältävien aineiden kanssa. Lääkeyhdisteissä pyritään välttämään vaikeaa erittymistä.




## Mikä on hyvä lääke?

Kysymys on ollut hyvin kokeellinen lääkekehityksessä kautta aikojen, ja kokeellisuudella pärjätään vieläkin pisimmälle. Ihmiset ovat esimerkiksi pajujen kaarnaa pureskelemalla tai Sechuanin pippureita maistelemalla saaneet kivunlievitystä erilaisiin vaivoihin. Antiobiootit löytyivät lähes sattumalta, mutta kokeilemalla niitä on löydetty uusia. Ihmiskunta tuntee jo tuhansia lääkkeitä, ja jälleen kerran informaatiotietieden hengessä keminformaatikkoja ja lääkäreitä on kiinnostanut: mikä niitä yhdistää?

Tähän kysymykseen Lipinksin johtama tutkimusryhmä etsi vastausta aloittaen suuren metatutkimuksen erilaisista patentoiduista lääkkeistä:

### Lipinskin viiden sääntö

Christopher Lipinski tutki tiiminsä kanssa n. 2000 lääkkeen dataa ja osoitti, että niillä on yleisesti päteviä hyviä piirteitä.



Lukuina sääntö tarkoittaa: 

- Molekyylimassa on alle 500 Da, jossa 1 Da vastaa noin 1 atomimassayksikkö 
- Jakautumiskerroin logP on alle 5 
- Vetydonorien lukumäärä on alle 5, jossa vetydonori on atomi/rakenne, joka vetysidoksen luovuttuaan toimii negatiivisella osittaisvarauksella (Esim. happi, O vesimolekyylien välisissä vetysidoksissa)
- Vetyakseptorien lukumäärä on alle 10, eli vetydonorien vastakohta siis pos. osittaisvarauksellinen atomi

Kuten huomaat. Näitä sääntöjä yhdistää se fakta, että kaikki luvut ovat jaettavissa numerolla 5. Tästä tulee nimitys: **Lipinskin viiden sääntö**.


Saatiin selville että useimmat lääkkeet poikkesivat vain vähän näistä arvoista korkealla luotettavuudella. Toisaalta Lipinski ym. eivät ottaneet huomioon luonnonyhdisteitä, ja moni luonnonyhdiste on todettu erittäin hyödylliseksi ja Lipisnkin sääntöjä rikkovaksi.


#### **Ekstra**: Lipinski ei ole aina oikeassa

Nykyään tieteilijät myös kritisoivat Lipinskin viiden sääntöä siitä, että se painottaa jokaista sääntöä yhtä suuresti.

*Säännöstä poikkeavaa ajattelua on olemassa myös sen takia, että jotkin yhdisteet toimivat loistavina lääkkeinä, vaikka ne rikkovat Lipinskin viiden sääntöä.*

**Daptomysiini**: Daptomysiini on lipopeptidipohjainen antibiootti, eli rasvaliukoinen proteiini, jota käytetään vakavien bakteeri-infektioiden hoitoon. Sen molekyylipaino on suhteellisen suuri (yli 1000 Da) ja sen korkean LogP-arvo tekee siitä erittäin rasvaliukoisen. Näistä ominaisuuksista huolimatta daptomysiini on tehokas lääke, koska se vaikuttaa metabolisesti tulehtuneessa kudoksessa, missä se pystyy tunkeutumaan bakteerisolukalvoihin. Kalvot menevät rakennemuutoksista lopulta rikki tuhoamalla bakteerin.

**Vankomysiini**: Vankomysiini on toinen antibiootti poikkeaa Lipinskin säännöistä sen massan ja rasvaliukoisuuden takia. Tosin tämän takia se pitää injektoida suoneen. Se on glykoproteiiniantibiootti, jota käytetään Gram-positiivisten bakteerien aiheuttamien infektioiden hoitoon. Vankomysiini toimii estämällä soluseinän synteesin bakteereissa, eli estäen bakteerien populaation kasvua. 

**Siklosporiini**: Siklosporiini on oraalisesti tai suonensisäisesti annosteltava immunosuppressiivinen lääke, jota käytetään elinsiirtojen jälkeisen käänteishyljinnän (keho hyökkää siirrännäistä vastaan) estolääkeenä sekä joissain autoimmuunisairauksissa, kuten nivelreumassa ja psoriasiksessa. Suuri molekyylipaino ja LogP-arvo, haittaavat lääkeaineen imeytymistä, mutta tähän usein pyritään vaikuttamaan niin, että injektioneste sisältää lääkeainetta emulsiona (vertaa maitoon).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Lataa dataa .csv tiedostosta taulukoksi
df = pd.read_csv('BP.csv', encoding='latin-1')

# Sovita lineaarinen regressiomalli 
model = LinearRegression()
X = df[['Moolimassa']]  # Feature (independent variable)
y = df['Kiehumispiste_K']  # Target variable (dependent variable)
model.fit(X, y)

# Piirrä  pistekuvio datasta 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Moolimassa', y='Kiehumispiste_K', data=df, color='blue', label='Data Points')

# Piirrä regressiosuora
plt.plot(X, model.predict(X), color='red', label='Linear Regression')

# Aseta otsikot ja mitat
plt.xlabel('Moolimassa')
plt.ylabel('Kiehumispiste (K)')
plt.title('Linear Regression Model')

# Show legend
plt.legend()

# Näytä sovitus
plt.show()
```
